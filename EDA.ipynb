How will you handle missing values?
○ You can replace them by calculating the average of the remaining
value of that column.
○ Will this be a good strate

It will depend which column we are considering, for example, for column price, filling the null values of price according to the average is not somethimg good, we can adopt other strategy such as grouping by in which we are filling the null values according to other features too because in real and practical life, many things are depending on other features too. In the given example, the price is depending on Stories, Bedrooms, Areas and Bathrooms etc. So 
what we are doing is that, that we are first categorizing the Houses according to the no.of bedrooms, no.of bathrooms, and no. of stories and then finding the mean of those particular price values.
In some cases, we will using ffill,bfill, interpolation etc.gy?
